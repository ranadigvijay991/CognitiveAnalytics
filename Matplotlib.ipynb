```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Boston.csv')
```

```python
plt.figure(figsize=(10,5))
plt.plot(df['age'])
plt.title('PLOT')
plt.xlabel('AGE')
plt.ylabel('INDICES')
```

```python
X = [1,3,5,7,9]
Y = [10,12,13,16,18]
plt.bar(X,Y)
```

```python
X1 = [1,3,5,7,9]
Y1 = [10,12,13,16,18]
X2 = [2,4,6,8,10]
Y2 = [11,16,14,12,15]

plt.figure(figsize=(10,5))
plt.bar(X1 , Y1 , color = 'cyan' , label = 'label1')
plt.bar(X2 , Y2 , color = 'magenta' , label = 'label2')
plt.legend()

plt.savefig('bar_fig.jpg')
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x = 'sepal_length' , y = 'petal_length' , data = data)
plt.show()
```
